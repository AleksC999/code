```python
# ОБЪЕДИНИТЬ ЛИСТЫ КНИГИ EXCEL
# Примененные функции
# pandas.ExcelFile
# pandas.read_excel
# list.append
# pandas.concat
# pandas.DataFrame.to_excel

## Сначала импортируем модули
import pandas as pd
```

```python
## Создаем объект ExcelFile.
xlsx = pd.ExcelFile('c:\\Users\\Alexk\\Documents\\Att\\Приложение 7_27-31.05.2024.xlsx')
```

```python
## Получаем список листов
worksheets = xlsx.sheet_names
```

```python
worksheets
```

```python
worksheets.remove('Справочник домов')
```

```python
worksheets
```

```python
## Создаем список, в котором каждый элемент — это датафрейм из одного листа.

worksheets_dfs = []

for worksheet in worksheets:
    data = pd.read_excel(xlsx,
                        skiprows=4,
                        sheet_name=worksheet)
    data['worksheet'] = worksheet
    worksheets_dfs.append(data)
```

```python
## Объединить все датафреймы из списка
worksheets_data = pd.concat(worksheets_dfs)
```

```python
worksheets_data
```

```python
worksheets_data.dtypes
```

```python
## Записать результат в книгу Excel.
worksheets_data.to_excel('c:\\Users\\Alexk\\Documents\\Att\\merger_.xlsx',
                         index=False)
```
